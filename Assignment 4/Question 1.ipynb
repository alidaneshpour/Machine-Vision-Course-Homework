# 1.Importing libraries

```python
#Importing Libraries
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np

#Necessary functions
def create_panorama(img1, img2):
    '''
    Combines two images into a single panorama.
    
    Parameters:
        img1 (numpy.ndarray): The first input image.
        img2 (numpy.ndarray): The second input image.
        
    Returns:
        numpy.ndarray: The resulting panorama image.
    '''
    
    panorama = None
    images = [img1, img2]
    stitcher = cv2.Stitcher_create()
    (status, panorama) = stitcher.stitch(images)
    
    return panorama

# Load two images
image1 = cv2.imread('image 1.jpg')
image2 = cv2.imread('image 2.jpg')

# Create the panorama
resulting_panorama = create_panorama(image1, image2)

#Show the output
plt.figure(figsize=(25, 16))
plt.imshow(cv2.cvtColor(resulting_panorama, cv2.COLOR_BGR2RGB))
plt.axis('off')  
plt.title('Panorama Result', fontsize=30)
plt.show()
```
